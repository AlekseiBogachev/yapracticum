## Описание проекта
Для исследования предоставлены данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктов за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.

### Предобработка данных

#### Обработка пропусков

Данные содержат пропуски, в некоторых столбцах больше половины пропущенных значений.

```python
(data
 .isna()
 .agg(['sum', 'mean'])
 .transpose()
 .sort_values('mean', ascending=False)
 .rename(columns={'sum':'number_of_missing_values',
                  'mean': 'fraction_of_missing_values'})
)
```

##### `locality_name`, `floors_total`, `living_area`, `kitchen_area`
Данные столбцы содержат относительно немного пропусков:
- `locality_name` - 0.2%
- `floors_total` - 0.4%
- `living_area` - 8.0%
- `kitchen_area` - 10.0%

Построим диаграммы размаха и гистограммы для всех столбцов с числовыми значениями, чтобы оценить наличие аномалий в данных. Графики будем выводить в алфавитном порядке названий столбцов.

```python
def box_and_hist_plot(data, column_name, bins=30):
    '''Выводит на экран диаграмму размаха и гистограмму на соседних графиках.'''
    
    fig, (ax0, ax1) = plt.subplots(1, 2, 
                                   figsize=(15,5), 
                                   gridspec_kw=dict(width_ratios=[1, 3])
                                  )
    
    data.plot(y=column_name, kind='box', ax=ax0, grid=True)
    data.plot(y=column_name, kind='hist', ax=ax1, bins=bins, grid=True)
    
    plt.suptitle(column_name)
    plt.show()


for col in data.select_dtypes(include=['int', 'float']).sort_index(axis='columns'):
    box_and_hist_plot(data, col)
```

Форма распределений в основном отличается от нормальной, и на диаграммах размаха отмечено довольно большое количество выбросов.

Удалим выбросы в следующих столбцах:
- `airports_nearest`,
- `ceiling_height`,
- `cityCenters_nearest`,
- `days_exposition`,
- `floor`,
- `floors_total`,
- `kitchen_area`,
- `last_price`,
- `living_area`,
- `parks_nearest`,
- `ponds_nearest`,
- `total_area`,
- `rooms`,
- `total_images`.

Эти столбцы выбранны, так как в них содержатся либо непрерывные величины, либо дискретные, содержащие достаточно большое количество уникальных значений, т.е. частью значений можно пожертвовать.

Сначала, вычислим границы каждого атрибута максимальным допустимым значением (`rightmost`) будет считаться значение соответствующее 0.995-му квантилю, минимальноым допустимым (`leftmost`) - значение, соответствующее 0.005-му квантилю.

Запишим значения границ в переменную `borders`.

```python
def rightmost(data):
    '''Вычисляет максимальное допустимое значение атрибута.'''
    return data.quantile(0.995)

def leftmost(data):
    '''Вычисляет минимальное допустимое значение атрибута.'''
    return data.quantile(0.005)

cols = ['airports_nearest',
        'ceiling_height',
        'cityCenters_nearest',
        'days_exposition',
        'floor',
        'floors_total',
        'kitchen_area',
        'last_price',
        'living_area',
        'parks_nearest',
        'ponds_nearest',
        'total_area',
        'rooms',
        'total_images',
       ]

borders = data[cols].agg([leftmost, rightmost]).transpose()
borders
```

Добавим к каждой строке датасета флаг `outlier`, обозначающий, что значение одного из атрибутов вышло за допустимые границы.

```python
data['outlier'] = False
```

```python
for col in cols:
    data['outlier'] |= (data[col] < borders.loc[col, 'leftmost']) | (data[col] > borders.loc[col, 'rightmost'])
```

Вычислим долю значений, отмеченных как выброс.

```python
print(f'Доля значений, отмеченных как выброс: {data.outlier.mean():.1%}')
```

Доля выбросов незначительна, мы можем удалить строки, в которых были обнаружены выбросы, затем удалить столбец `outlier`.

```python
data.drop(index=data.query('outlier').index, inplace=True)
```

##### Связь полной стоимости и общей площади
Построим диаграмму рассеяния и вычислим коэффициент корреляции.

```python
def discover_correlation(col_name):
    '''Выводит диаграмму рассеяния и вычисляет корреляцию для указанного столбцов с именами col_name и "last_price".'''
    data.plot(x=col_name, y='last_price', kind='scatter', grid=True, alpha=0.1)
    plt.show()

    print(f"Коэффициент корреляции равен {data.corr().loc[col_name, 'last_price']:.2f}")
    
discover_correlation('total_area')
```

Таким образом, между полной стоимостью и общей площадью наблюдается довольно сильная корреляция.

##### Связь полной стоимости и жилой площади

Корреляция между полной стоимостью и жилой площадью существенна, но меньше, чем между общей площадью и полной стоимостью.

#### Цена одного квадратного метра в 10 населённых пунктах с наибольшим числом объявлений
Посчитаем среднюю цену одного квадратного метра в 10 населённых пунктах с наибольшим числом объявлений.

```python
top_localities = list(data.locality_name.value_counts(ascending=False)[:10].index)
top_price_per_square_meter = (data[data.locality_name.isin(top_localities)]
                              .groupby('locality_name')
                              .price_per_square_meter
                              .mean()
                              .sort_values(ascending=False)
                              .round(2)
                             )

print('Цена одного квадратного метра в 10 населённых пунктах с наибольшим числом объявлений (руб.):')
print(top_price_per_square_meter)
print()

top_price_per_square_meter.plot(y='price_per_square_meter', 
                                kind='bar',
                                xlabel='Город',
                                ylabel='Средняя цена одного кв.м, р.',
                                title='Средняя цена одного кв.м в городах\nс самым большим количеством объявлений',
                                grid=True,
                                figsize=(10, 7)
                               )

print(f'Самая высокая стоимость квадратного метра в городе {top_price_per_square_meter.index[0]}')
print(f'Самая низкая стоимость квадратного метра в городе {top_price_per_square_meter.index[-1]}')
```
